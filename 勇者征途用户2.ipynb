```python
import pandas as pd
import json
```

```python
#去除重复值
#id 数据类型转换
df = pd.read_csv('勇者征途用户2.csv')
df.columns = ['id','game_type','user_json']
df
#df['id'].value_counts()
df['id'] = df['id'].drop_duplicates()
df['id'].value_counts()
df = df.dropna()
df['id'] = df['id'].astype('int')
df['id'] = df['id'].astype('str')
df.head
```

```python
temp = [json.loads(i) for i in df.user_json]
temp[:5]
tempdf = pd.DataFrame.from_dict(temp)
df = df.drop('user_json', axis = 1).merge(tempdf)
df.head()
```

```python
cour_list= []
file = df['user_json']
for lines in file:
    cour_lines = json.loads(lines)
    cour_lines
cour_lines
```
